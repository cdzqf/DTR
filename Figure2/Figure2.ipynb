### 1. BA scale-free networks

#### Basic structural characteristics of BA scale-free networks in the DTR flow, the parameter $m = 2$.

```python
KT = [0, 2, 4, 6] # m = 2
N = 10000
m = 2

G = nx.barabasi_albert_graph(N, m)
x01, x02, x03, y01, y02, y03 = cal_three_characteristics(G)

G0 = G.copy()
G1 = DTR(G0, KT[1])
N1 = nx.number_of_nodes(G1)
x11, x12, x13, y11, y12, y13 = cal_three_characteristics(G1)

G0 = G.copy()
G2 = DTR(G0, KT[2])
N2 = nx.number_of_nodes(G2)
x21, x22, x23, y21, y22, y23 = cal_three_characteristics(G2)

G0 = G.copy()
G3 = DTR(G0, KT[3])
N3 = nx.number_of_nodes(G3)
x31, x32, x33, y31, y32, y33 = cal_three_characteristics(G3)
```

```python
with plt.style.context(['science', 'scatter']):
    fig, axs = plt.subplots(1, 3, figsize=(9, 2.5))

    ax1 = axs[0]
    ax1.plot(x01, y01, clip_on=False, label="$k_T =$ "+str(KT[0])+", $N(k_T) =$ "+str(N))
    ax1.plot(x11, y11, clip_on=False, label="$k_T =$ "+str(KT[1])+", $N(k_T) =$ "+str(N1))
    ax1.plot(x21, y21, clip_on=False, label="$k_T =$ "+str(KT[2])+", $N(k_T) =$ "+str(N2))
    ax1.plot(x31, y31, clip_on=False, label="$k_T =$ "+str(KT[3])+", $N(k_T) =$ "+str(N3))
    ax1.legend(loc=3, edgecolor = "black", handlelength=0.2, fontsize=8, frameon=True, fancybox=False)
    ax1.set_xlabel(r"$k'$")
    ax1.set_ylabel(r"$P_c(k')$")
    ax1.set_xscale("log")
    ax1.set_yscale("log")
    ax1.set_ylim(5e-5, 1)
    ax1.set_title(r'\textbf{(a)}', x = 0.9, y = 0.85)


    ax2 = axs[1]
    ax2.plot(x02, y02, clip_on=False)
    ax2.plot(x12, y12, clip_on=False)
    ax2.plot(x22, y22, clip_on=False)
    ax2.plot(x32, y32, clip_on=False)
    ax2.set_xlabel(r"$k'$")
    ax2.set_ylabel(r"$\bar c(k')$")
    ax2.set_xscale("log")
    ax2.set_yscale("log")
    ax2.set_ylim(1e-4, 1)
    ax2.set_title(r'\textbf{(b)}', x = 0.9, y = 0.85)


    ax3 = axs[2]
    ax3.plot(x03, y03, clip_on=False)
    ax3.plot(x13, y13, clip_on=False)
    ax3.plot(x23, y23, clip_on=False)
    ax3.plot(x33, y33, clip_on=False)
    ax3.set_xlabel(r"$k'$")
    ax3.set_ylabel(r"$\bar k_{nn}(k')$")
    ax3.set_xscale("log")
    ax3.set_yscale("log")
    ax3.set_ylim(1e-1, 1e1)
    ax3.set_title(r'\textbf{(c)}', x = 0.9, y = 0.85)

    plt.subplots_adjust(left=0.065, bottom=0.20, right=0.98, top=0.91, wspace=0.35)
    # plt.savefig("Figure2.pdf")
    plt.show()
```
